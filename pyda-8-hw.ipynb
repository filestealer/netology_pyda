Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
grouped = ratings.groupby('userId').count()
grouped.head()
users_m100 = grouped[grouped['rating'] > 100].reset_index()['userId'].tolist()
```

```python
agg_timestamp = (ratings[ratings['userId'].isin(users_m100)]
                 .groupby('userId')['timestamp'].agg([min, max])
                )
```

```python
(agg_timestamp['max']-agg_timestamp['min']).mean()
```

Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). Необходимо сформировать две таблицы:

таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
аналогичную таблицу по типам выручки с указанием адреса клиента
Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
# К домашнему заданию, задача 2
# Дана статистика услуг перевозок клиентов компании по типам:

# rzd - железнодорожные перевозки
# auto - автомобильные перевозки
# air - воздушные перевозки
# client_base - адреса клиентов
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# client_base['client_id'].join(rzd, on='client_id', how='left') а что с столбцом делать?)
(client_base
     .merge(rzd, on='client_id', how='left')
     .merge(auto, on='client_id', how='left')
     .merge(air, on='client_id', how='left')
)[[ 'client_id', 'rzd_revenue', 'auto_revenue', 'air_revenue']]
```

```python
# client_base['client_id'].join(rzd, on='client_id', how='left') а что с столбцом делать?)
(client_base
     .merge(rzd, on='client_id', how='left')
     .merge(auto, on='client_id', how='left')
     .merge(air, on='client_id', how='left')
)
```

В задаче сквозной аналитики вам предоставили данные по местоположению пользователей. Т. е. для каждого user_id известна последовательность координат (широта/долгота), когда они требовались приложению для полноценной работы. Как бы вы добавили эти сведения в таблицу визитов и покупок?

```python
visits = pd.DataFrame(
    {
        'user_id': [11, 22, 55, 11, 77],
        'source': ['ad', 'yandex', 'email', 'google', 'ad']
    }
)

visits = visits[['user_id', 'source']]
visits
```

```python
visits_grouped = visits.groupby('user_id').count()
visits_grouped.rename(columns={'source': 'visits'}, inplace=True)
# visits_grouped = visits.groupby('user_id').count().reset_index()
visits_grouped
```

```python
purchases = pd.DataFrame(
    {
        'user_id': [11, 22, 55, 11, 99],
        'category': ['Спорт', 'Авто', 'Дача', 'Спорт', 'Авто'],
    }
)

purchases = purchases[['user_id', 'category']]
purchases
```

```python
locations = pd.DataFrame(
    {
        'user_id': [11, 22, 55, 11, 99],
        'longitude': [33, 33, 34, 35, 36],
        'latitude': [74, 74, 75, 77, 99],
    }
)

locations = locations[['user_id', 'longitude', 'latitude']]
locations
```

```python
purchases_pivot = purchases.pivot_table(index='user_id', columns='category', values='user_id', 
                                        aggfunc='size', fill_value=0)
# purchases_pivot = purchases_pivot.reset_index()
purchases_pivot
```

```python
visits_grouped.join(purchases_pivot).merge(locations, on='user_id', how='left')
```
